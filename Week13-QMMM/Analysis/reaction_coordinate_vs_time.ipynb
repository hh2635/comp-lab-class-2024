```python
# reaction_coordinate_analysis.ipynb

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define file paths
colvar_file = "../Outputs/monitor/MONITOR-COLVAR.metadynLog"
figures_dir = "../Figures"
os.makedirs(figures_dir, exist_ok=True)

# Parse the COLVAR file
def parse_colvar_file(file_path):
    """Parse COLVAR file to extract time and reaction coordinate."""
    time = []
    reaction_coordinate = []
    with open(file_path, "r") as file:
        for line in file:
            if line.startswith("#"):  # Skip comments
                continue
            data = line.split()
            time.append(float(data[0]) / 1000.0)  # Convert time to ps
            reaction_coordinate.append(float(data[1]))  # Reaction coordinate value
    return time, reaction_coordinate

# Parse the COLVAR file
time, reaction_coordinate = parse_colvar_file(colvar_file)

# Plot reaction coordinate vs. time
plt.figure(figsize=(10, 6))
plt.plot(time, reaction_coordinate, label="Reaction Coordinate", color="purple")
plt.xlabel("Time (ps)")
plt.ylabel("Reaction Coordinate (C-O Distance - C-C Distance)")
plt.title("Reaction Coordinate vs. Time")
plt.legend()
plt.grid()
plt.tight_layout()

# Save the plot
output_plot = os.path.join(figures_dir, "reaction_coordinate_vs_time.png")
plt.savefig(output_plot)
plt.show()

# Print completion message
print(f"Plot saved to: {output_plot}")
```
