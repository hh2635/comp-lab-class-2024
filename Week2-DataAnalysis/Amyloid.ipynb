```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load the MD trajectory file
# Replace 'trajectory.xtc' and 'topology.pdb' with the actual trajectory and topology file paths
trajectory = md.load('Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='Data/1hz3_T310.start.nowater.gro')

# Step 2: Compute the end-to-end distance at each time step
# The end-to-end distance is the distance between the first and last residue's alpha-carbon (CA)
atom_indices = trajectory.topology.select('name CA')
end_to_end_distances = md.compute_distances(trajectory, [[atom_indices[0], atom_indices[-1]]])

# Step 3: Compute the radius of gyration at each time step
radii_of_gyration = md.compute_rg(trajectory)

# Step 4: Plot end-to-end distance and radius of gyration vs time
time = trajectory.time / 1000  # Convert time to nanoseconds for plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(time, end_to_end_distances, label='End-to-End Distance (nm)', color='blue', linestyle='--')
ax.plot(time, radii_of_gyration, label='Radius of Gyration (nm)', color='red', linestyle='-')
ax.set_xlabel('Time (ns)')
ax.set_ylabel('Distance (nm)')
ax.set_title('End-to-End Distance and Radius of Gyration vs Time')
ax.legend()
plt.show()

# Step 5: Plot a normalized histogram of end-to-end distance
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].hist(end_to_end_distances, bins=30, density=True, color='blue', alpha=0.7)
ax[0].set_xlabel('End-to-End Distance (nm)')
ax[0].set_ylabel('Probability Density')
ax[0].set_title('Histogram of End-to-End Distance')

# Step 6: Plot a normalized histogram of radius of gyration
ax[1].hist(radii_of_gyration, bins=30, density=True, color='red', alpha=0.7)
ax[1].set_xlabel('Radius of Gyration (nm)')
ax[1].set_ylabel('Probability Density')
ax[1].set_title('Histogram of Radius of Gyration')

plt.tight_layout()
plt.show()
```
