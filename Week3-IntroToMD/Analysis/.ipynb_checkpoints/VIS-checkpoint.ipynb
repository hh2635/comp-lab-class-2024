```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

## Energy Minimization

```python
# Path to the .xvg file
Ene_min_path = "/home/hh2635/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Data/potential.xvg"

# Step 1: Read the file and filter out comment lines
with open(Ene_min_path, 'r') as file:
    lines = file.readlines()
    
# Filter out lines that start with '#' or '@'
filtered_lines = [line for line in lines if not line.startswith(('#', '@'))]

# Join filtered lines and use Pandas to read the data
from io import StringIO
data = '\n'.join(filtered_lines)

# Load the data into a DataFrame
df = pd.read_csv(StringIO(data), delim_whitespace=True, header=None)
print(df.head()) 
```

```python
df.columns = ["0", "1"]

# Plot the functions
plt.figure(figsize=(10, 8))
# plt.xlim(-10, 10)

plt.plot(df["0"], df["1"], label='potential', color='black', linewidth=2.5)


# Labels and title
plt.xlabel('Energy Minimizing Step', fontsize=16, fontweight="bold")
plt.ylabel('energy (kJ/mol)', fontsize=16, fontweight="bold")
plt.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))
plt.title('Energy Minimization of 2JOF', fontsize=16, fontweight="bold")
plt.tick_params(axis='both', direction='in', labelsize=12, width=2.5)
plt.xticks(fontsize="16", fontweight='bold')  # Bold x-axis numbers
plt.yticks(fontsize="16", fontweight='bold')  # Bold y-axis numbers


# Bold the spines (borders)
for spine in plt.gca().spines.values():
    spine.set_linewidth(2.5)  # Set the width of the spines

# plt.axhline(0, color='black', linewidth=0.5)
# plt.axvline(0, color='black', linewidth=0.5)
# plt.grid(True)
plt.legend(fontsize=12, frameon=True, loc='upper right')

# Show the plot
plt.show()
```

```python
# Path to the .xvg file
Pres_equiv_path = "/home/hh2635/comp-lab-class/comp-lab-class-2024/Week3-IntroToMD/Data/pressure.xvg"

# Step 1: Read the file and filter out comment lines
with open(Pres_equiv_path, 'r') as file:
    lines = file.readlines()
    
# Filter out lines that start with '#' or '@'
filtered_lines = [line for line in lines if not line.startswith(('#', '@'))]

# Join filtered lines and use Pandas to read the data
from io import StringIO
data = '\n'.join(filtered_lines)

# Load the data into a DataFrame
df = pd.read_csv(StringIO(data), delim_whitespace=True, header=None)
print(df.head()) 
```

```python
df.columns = ["0", "1"] # len=101

Pres_array = np.array(df["1"])

# Function to calculate the moving average with a window size of 10
def moving_average(arr, window_size):
    return np.convolve(arr, np.ones(window_size) / window_size, mode='valid')

# Calculate the moving average for every 10 consecutive numbers
window_size = 10
avg_10step = moving_average(Pres_array, window_size)
# print(type(avg_10step))
nan_index = np.full(9, np.nan)
avg_press = np.concatenate((nan_index, avg_10step))
print(len(avg_press))
```

```python
# Plot the functions
plt.figure(figsize=(10, 8))
plt.xlim(0, 100)

plt.plot(df["0"], df["1"], label='Pressure', color='black', linewidth=2.5)
plt.plot(df["0"], avg_press, label='10-ps running avg', color='red', linewidth=2.0)


# Labels and title
plt.xlabel('Time (ps)', fontsize=16, fontweight="bold")
plt.ylabel('Pressure (bar)', fontsize=16, fontweight="bold")
# plt.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))
plt.title('Pressure at NVT Equilibrium', fontsize=16, fontweight="bold")
plt.tick_params(axis='both', direction='in', labelsize=12, width=2.5)
plt.xticks(fontsize="16", fontweight='bold')  # Bold x-axis numbers
plt.yticks(fontsize="16", fontweight='bold')  # Bold y-axis numbers


# Bold the spines (borders)
for spine in plt.gca().spines.values():
    spine.set_linewidth(2.5)  # Set the width of the spines

# plt.axhline(0, color='black', linewidth=0.5)
# plt.axvline(0, color='black', linewidth=0.5)
# plt.grid(True)
plt.legend(fontsize=12, frameon=True, loc='upper right')

# Show the plot
plt.show()
```
