```python
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the COLVAR file using numpy
# Replace 'COLVAR' with the correct path to your file
data = np.genfromtxt('COLVAR', comments='#')

# Extract phi and psi
phi = data[:, 1]
psi = data[:, 2]

# Create the Ramachandran-like plot
plt.figure(figsize=(6, 6))
plt.scatter(phi, psi, c='blue', alpha=0.6)

# Set axis labels and title
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('Ramachandran-like plot (Phi vs Psi)')

# # Set axis limits to reflect the range of phi and psi (-pi to pi)
# plt.xlim([-np.pi, np.pi])
# plt.ylim([-np.pi, np.pi])

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np

# Load the COLVAR file
data = np.genfromtxt('COLVAR', comments='#')

# Extract phi and psi columns
phi = data[:, 1]
psi = data[:, 2]

# Compute the standard deviations
std_phi = np.std(phi)
std_psi = np.std(psi)

print(f"Standard deviation of phi: {std_phi}")
print(f"Standard deviation of psi: {std_psi}")
```
