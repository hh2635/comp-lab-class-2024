```python
import numpy as np
import matplotlib.pyplot as plt

# Load the FES grid file (assuming it contains phi, psi, and free energy columns)
fes_data = np.genfromtxt('fes_grid.dat')

# Extract phi, psi, and free energy values
phi = fes_data[:, 0]
psi = fes_data[:, 1]
free_energy = fes_data[:, 2]

# Reshape the data to form a grid
n_points = int(np.sqrt(len(fes_data)))  # Assuming a square grid
phi_grid = phi.reshape((n_points, n_points))
psi_grid = psi.reshape((n_points, n_points))
free_energy_grid = free_energy.reshape((n_points, n_points))

# Plot the free energy surface using imshow and add contour lines
plt.figure(figsize=(8, 6))
plt.imshow(free_energy_grid, extent=[-np.pi, np.pi, -np.pi, np.pi], origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label='Free Energy (kJ/mol)')
plt.contour(phi_grid, psi_grid, free_energy_grid, levels=10, colors='white', linewidths=0.5)

# Label the axes
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('2D Free Energy Surface (FES) for Phi and Psi')

# Show the plot
plt.show()
```
