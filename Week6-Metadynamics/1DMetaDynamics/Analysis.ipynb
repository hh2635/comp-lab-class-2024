```python
import numpy as np
import matplotlib.pyplot as plt

# Load the COLVAR file
# Assuming columns: time, phi, psi, bias
data = np.genfromtxt('COLVAR', comments='#')

# Extract time and phi values
time = data[:, 0]  # Time is in the first column
phi = data[:, 1]   # Phi is in the second column

# Plot phi vs time
plt.plot(time, phi, label='Phi (radians)', color='blue')
plt.xlabel('Time (ps)')
plt.ylabel('Phi (radians)')
plt.title('Phi vs Time during Metadynamics Simulation')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the HILLS file, assuming columns: time, phi, sigma_phi, height, biasf
data = np.genfromtxt('HILLS', comments='#')

# Extract time and height columns
time = data[:, 0]   # Time is in the first column
height = data[:, 3] # Gaussian height is in the fourth column

# Plot the Gaussian height vs time
plt.plot(time, height, label='Gaussian Height', color='green')
plt.xlabel('Time (ps)')
plt.ylabel('Height (kJ/mol)')
plt.title('Gaussian Height vs Time during Metadynamics')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the phi_grid.dat file (assuming the columns are phi and bias energy)
data = np.genfromtxt('phi_grid.dat')

# Extract phi values and bias energy
phi = data[:, 0]  # Assuming phi is the first column
bias_energy = data[:, 1]  # Assuming bias energy is the second column

# Plot the negative of the bias energy (approximate FES)
plt.plot(phi, -bias_energy)
plt.xlabel('Phi (radians)')
plt.ylabel('Free Energy (kJ/mol)')
plt.title('Free Energy Surface (FES) from Metadynamics')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Manually select the FES files corresponding to the times you want
# Based on the files you have, choose the ones closest to 0.1 ns, 0.3 ns, 1 ns, 2 ns, 5 ns, and 10 ns
selected_files = [
    'fes_0.dat',  # Closest to 0.1 ns
    'fes_3.dat',  # Closest to 0.3 ns
    'fes_10.dat', # Closest to 1 ns
    'fes_20.dat', # Closest to 2 ns
    'fes_50.dat', # Closest to 5 ns
    'fes_100.dat' # Closest to 10 ns
]

# Times associated with each file
times = [0.1, 0.3, 1, 2, 5, 10]

# Prepare the plot
plt.figure(figsize=(8, 6))

# Loop over the selected FES files and plot the free energy profile
for fes_file, time in zip(selected_files, times):
    try:
        # Load the data from the current FES file
        data = np.genfromtxt(fes_file)
        
        # Extract phi and free energy columns
        phi = data[:, 0]         # Phi values
        free_energy = data[:, 1]  # Free energy values

        # Plot the free energy profile for this file
        plt.plot(phi, free_energy, label=f'{time} ns', alpha=0.7)
    
    except FileNotFoundError:
        print(f"File {fes_file} not found, skipping this time point.")

# Customize the plot
plt.xlabel('Phi (radians)')
plt.ylabel('Free Energy (kJ/mol)')
plt.title('Free Energy Surface (FES) Profiles Over Time')
plt.grid(True)
plt.legend(loc='best')

# Show the plot
plt.show()
```
